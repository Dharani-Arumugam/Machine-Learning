## Telecom Churn Case Study
With 21 predictor variables we need to predict whether a particular customer will switch to another telecom provider or not. In telecom terminology, this is referred to as churning and not churning, respectively.

### 1. Importing and Merging Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
# read churn data
churn_data    = pd.read_csv('churn_data.csv')
churn_data.head(5)
```

```python
# read customer_data
customer_data = pd.read_csv('customer_data.csv')
customer_data.head(5)
```

```python
# read internet data
internet_data = pd.read_csv('internet_data.csv')
internet_data.head(5)
```

```python
## Merge all dataframe into one using cutomer_id
df_1    = pd.merge(churn_data, customer_data,how ='inner', on ='customerID')
telecom = pd.merge(df_1, internet_data, how ='inner', on ='customerID')
telecom.head(5)
```

```python
telecom.info()
```

- All dataframes are merged to get the new dataframe with 21 columns

```python
telecom.describe()
```

- Among 21, only 3 columns are numerical columns 

### Data Preperation

#### Convert the binary variables `(yes/no)` into `0/1 `

```python
binary_cols = ['PhoneService','PaperlessBilling','Partner','Dependents','Churn']

telecom[binary_cols] = telecom[binary_cols].apply(lambda x: x.map({'Yes':1,'No':0}))

telecom[binary_cols].head(5)
```

```python
## Creating dummy variables for columns with multiple categories and drop the first one
dummies = pd.get_dummies(telecom[['Contract','PaymentMethod','gender','InternetService']], drop_first=True)

telecom = pd.concat([telecom,dummies],axis=1)
```

```python
telecom.head(5)
```

#### Creating dummy variables for remaining categorical variables and drop the columns with max col name length 

```python
## MultipleLines column
m1      = pd.get_dummies(telecom['MultipleLines'],prefix='MultipleLines')
m1      = m1.drop('MultipleLines_No phone service', axis=1)
telecom = pd.concat([telecom, m1], axis=1)
```

```python
## OnlineSecurity column
os      = pd.get_dummies(telecom['OnlineSecurity'],prefix='OnlineSecurity')
os      = os.drop('OnlineSecurity_No internet service', axis=1)
telecom = pd.concat([telecom, os], axis=1)
```

```python
## OnlineBackup column
ob      = pd.get_dummies(telecom['OnlineBackup'],prefix='OnlineBackup')
ob      = ob.drop('OnlineBackup_No internet service',axis=1)
telecom = pd.concat([telecom,ob], axis=1)
```

```python
## DeviceProtection column
dp      = pd.get_dummies(telecom['DeviceProtection'],prefix='DeviceProtection')
dp      = dp.drop('DeviceProtection_No internet service',axis=1)
telecom = pd.concat([telecom,dp], axis=1)
```

```python
## TechSupport column
ts      = pd.get_dummies(telecom['TechSupport'], prefix='TechSupport')
ts      = ts.drop('TechSupport_No internet service',axis=1)
telecom = pd.concat([telecom,ts],axis=1)
```

```python
## StreamingTV
st      = pd.get_dummies(telecom['StreamingTV'],prefix ='StreamingTV')
st      = st.drop('StreamingTV_No internet service', axis=1)
telecom = pd.concat([telecom,st],axis=1)
```

```python
## StreamingMovies
sm      = pd.get_dummies(telecom['StreamingMovies'],prefix='StreamingMovies')
sm      = sm.drop('StreamingMovies_No internet service',axis=1)
telecom = pd.concat([telecom,sm],axis=1)
```

```python
telecom.head(5)
```

### Dropping Repeated Columns

```python
# We have created dummies for the below variables, so we can drop them
telecom = telecom.drop(['Contract','PaymentMethod','gender','MultipleLines','InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies'], 1)
```

```python
## Convert TotalCharges column to a numeric type
telecom['TotalCharges'] = pd.to_numeric(telecom['TotalCharges'],errors='coerce')
```

```python
telecom.info()
```

### Checking Outliers

```python
num_cols = ['tenure','MonthlyCharges','TotalCharges']

telecom[num_cols].describe(percentiles=[.25,.5,.75,.90,.95,.99])
```

- As seen, the numbers doesn't show any outliers

### Handling Missing values

```python
# Check percentage of missing values in each column
round(100*(telecom.isnull().sum()/len(telecom.index)),2)
```

- Only TotalCharges column has 0.16% null values.It is better to remove them

```python
## Remove the TotalCharges column with null values 
telecom = telecom[~(telecom['TotalCharges'].isnull())]
```

```python
# Check percentage of missing values in each column
round(100*(telecom.isnull().sum()/len(telecom.index)),2)
```

- No missing data is there in the dataframe

### Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X = telecom.drop(['Churn','customerID'],axis=1)
X.head(5)
```

```python
y = telecom['Churn']
y.head(5)
```

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

- Among all the variables, `tenure`, `MonthlyCharges`,`TotalCharges` are columns with continuos values. Hence scaling can be done on these columns

```python
scaler = StandardScaler()
X_train[['tenure','MonthlyCharges','TotalCharges']] = scaler.fit_transform(X_train[['tenure','MonthlyCharges','TotalCharges']])
X_train.head(5)
```

- All numerical columns are now transformed to a scale of 1 to -1

### Find Imbalance ratio

```python
100*(sum(telecom['Churn'])/len(telecom['Churn'].index))
```

- Churn ratio is about 27% which is ok to proceed

### Model Building

```python
import statsmodels.api as sm

#Logistic Regression model 
logm1 = sm.GLM(y_train, sm.add_constant(X_train), family= sm.families.Binomial())
logm1.fit().summary()
```

### Feature Selection using RFE

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
from sklearn.feature_selection import RFE 
rfe = RFE(logreg, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
rfe.support_
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
```

```python
# Columns which are removed from RFE
X_train.columns[~rfe.support_]
```

### Model Assessment with StatsModels

```python
X_train_sm = sm.add_constant(X_train[col])
logm2 = sm.GLM(y_train, X_train_sm, family =sm.families.Binomial())
res   = logm2.fit()
res.summary()
```

- Most of the p-values are less and so it's consistent

```python
y_train_pred = res.predict(X_train_sm)
y_train_pred.head(5)
```

```python
y_train_pred = y_train_pred.values.reshape(-1)
y_train_pred[:10]
```

#### Creating a dataframe with actual churn flag and predicted probabilities

```python
y_train_pred_final = pd.DataFrame({'Churn':y_train.values, 'Churn_Prob':y_train_pred})
y_train_pred_final['CustId'] = y_train.index
y_train_pred_final.head(5)
```

#### Creating a column 'Predicted'

```python
y_train_pred_final['Predicted'] = y_train_pred_final['Churn_Prob'].apply(lambda x: 1 if x > 0.5 else 0)
y_train_pred_final.head(5)
```

### Calculating confusion matrix to check accuracy

```python
from sklearn import metrics
```

```python
confusion_matrix = metrics.confusion_matrix(y_train_pred_final.Churn,y_train_pred_final.Predicted)
confusion_matrix
```

- Confustion matrix values are shown

```python
print(metrics.accuracy_score(y_train_pred_final.Churn, y_train_pred_final.Predicted))
```

- Accuracy score `81%`
